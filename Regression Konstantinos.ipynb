```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

from sklearn.linear_model import LinearRegression

import seaborn as sns

from IPython.display import display
pd.options.display.max_columns = None
```

```python
cars = pd.read_csv("vehicles_clean3.csv")
cars.head()
```

```python
cars = cars.set_index("vin")

cars["age_int_vintage"] = cars["vintage"] * cars["age"]

cars["age_squared"] = cars["age"] ** 2

cars.head()
```

```python
cars = cars[["price"
              ,"mean_price_by_model_name"
              ,"mileage"
              ,"horsepower"
              ,"age"
              ,"savings_amount"
              ,"wheelbase"
              ,"major_options_number"
              ,"fuel_type_Compressed Natural Gas"
              ,"fuel_type_Diesel"
              ,"fuel_type_Flex Fuel Vehicle"
              ,"fuel_type_Gasoline"
              ,"fuel_type_Hybrid"
              ,"fuel_type_Propane"
              ,"engine_type_H6"
              ,"engine_type_I2"
              ,"engine_type_I3"
              ,"engine_type_I4"
              ,"engine_type_I5"
              ,"engine_type_I6"
              ,"engine_type_R2"
              ,"engine_type_V10"
              ,"engine_type_V12"
              ,"engine_type_V6"
              ,"engine_type_V8"
              ,"engine_type_W12"
              ,"engine_type_W8"
              ,"body_type_Coupe"
              ,"body_type_Hatchback"
              ,"body_type_Minivan"
              ,"body_type_Pickup Truck"
              ,"body_type_SUV / Crossover"
              ,"body_type_Sedan"
              ,"body_type_Van"
              ,"body_type_Wagon"
              ,"age_squared"
              ,"age_int_vintage"]]

cars.head()
```

```python
cars_train, cars_test = train_test_split(cars, test_size = 0.2, random_state = 42)

X_train = cars_train.drop(["price"], axis = 1)
y_train = cars_train["price"]

X_test = cars_test.drop(["price"], axis = 1)
y_test = cars_test["price"]
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
rmse = mean_squared_error(y_pred, y_test, squared=False)
rmse
```

```python
r_squared = r2_score(y_pred, y_test)
r_squared
```
