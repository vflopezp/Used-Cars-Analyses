```python
import pandas as pd
from IPython.display import display
pd.options.display.max_columns = None
```

```python
df = pd.read_csv("vehicles_clean2.csv")
```

```python
mean_price_by_make_name = df.groupby('make_name')['price'].mean()
# mean_price_by_make_name.index
# mean_price_by_make_name['Acura'] --> outputs the mean price for all cars brand Acura

mean_price_by_make_name
```

```python
arr_mean_price_by_make_name = []

for brand in df['make_name']:
    arr_mean_price_by_make_name.append(round(mean_price_by_make_name[brand],2))

df['mean_price_by_make_name'] = arr_mean_price_by_make_name

df.head()
```

```python
mean_price_by_model_name = df.groupby('model_name')['price'].mean()

arr_mean_price_by_model_name = []

for model in df['model_name']:
    arr_mean_price_by_model_name.append(round(mean_price_by_model_name[model],2))

df['mean_price_by_model_name'] = arr_mean_price_by_model_name

df.head()
```
